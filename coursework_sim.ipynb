# CW6

```python
w = world.World()
w.draw_map()
w.draw_particles()
plt.imshow(w.display)
```

```python
import random
import landmarks as lm
from sklearn.metrics import confusion_matrix
import pandas as pd

y_true = []
y_pred = []
correct = 0
count = 0
lm_correct = 0
for i in range(100):
    landmark_placed = i%5
    angle_placed = random.random()*2*pi-pi
    x_placed, y_placed = lm.positions[landmark_placed]
    w = world.World(starting_x=x_placed,starting_y=y_placed,starting_theta=angle_placed)
    pred_lm,pred_angle = w.find_landmark_and_orientation()
    print("true landmark = "+str(landmark_placed)+", landmark predicted = "+str(pred_lm)+", true angle = "+str(angle_placed*180.0/pi)+", predicted angle = "+str(pred_angle*180.0/pi))
    print("-------------------------")
    y_true.append(landmark_placed)
    y_pred.append(pred_lm)
    if landmark_placed == pred_lm:
        lm_correct +=1
    if (abs(angle_placed-pred_angle)*180.0/pi < 3):
        correct += 1
    count += 1
print(">>> lmark accuracy = "+str(100.0*lm_correct/count)+" %")
print(">>> angle accuracy = "+str(100.0*correct/count)+" %")
```

```python
confusion = confusion_matrix(y_true,y_pred)
plt.figure(figsize = (7,5))
plt.matshow(confusion)
plt.colorbar()
```

# CW7

```python
#the robot is placed into a random landmark
landmark_placed = random.randint(0,4)
angle_placed = random.random()*2*pi-pi
x_placed, y_placed = lm.positions[landmark_placed]
w = world.World(starting_x=x_placed,starting_y=y_placed,starting_theta=angle_placed)

#it then finds the landmark and its orientation
pred_lm,pred_angle = w.find_landmark_and_orientation()
print("======== context =========")
print("true landmark = "+str(landmark_placed)+", landmark predicted = "+str(pred_lm)+", true angle = "+str(angle_placed*180.0/pi)+", predicted angle = "+str(pred_angle*180.0/pi))
print("==========================")

#the robot then plans its path
print("======== path planning =========")
waypoints = w.path_plan(pred_lm)
print(waypoints)
print("================================")

#then it goes to each landmark, hopefully
w.navigate_waypoints(waypoints)

print("========== end point ===========")
print(w.r)
print("================================")
```
